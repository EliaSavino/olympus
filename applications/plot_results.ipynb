```python
import pickle
import numpy as np
import pandas as pd

from scipy.stats import pearsonr, spearmanr

import matplotlib.pyplot as plt
import seaborn as sns

import olympus
```

```python
# load the raw data
df = pd.read_csv('../src/olympus/datasets/dataset_opv/screendata.csv')
print(df.shape)
print(df.columns)
df.head()
```

```python
columns = ['HOMO(eV)', 'LUMO(eV)', 'IP(eV)', 'EA(eV)', 'excE(eV)', 'exc_osc(a.u.)']
```

```python
# estimate the pearson correlation between the properties
for col1 in columns:
    for col2 in columns:
        corr = pearsonr(df[col1], df[col2])[0]
        print(col1, col2, corr)
```
